LINEAR REGRESSION

```python
lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
y_pred = lr.predict(X_test)
#y_test
```

```python
pd.DataFrame({
    'y_prediction': y_pred,
    'x' : y_test
})
```

```python
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)
print('Mean_Squared_Error :' ,mse)
print('r_square_value :',r_squared)
```
